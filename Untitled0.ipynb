```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from imblearn.combine import SMOTETomek
from sklearn.utils.class_weight import compute_class_weight

# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Display basic info
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
print("\nData types and null values:")
display(df.info())
print("\nClass distribution:")
print(df['Attrition'].value_counts())
```

```python
# Drop columns with single unique value or no predictive power
cols_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
df = df.drop(cols_to_drop, axis=1)

# Encode categorical variables
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Separate features and target
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nTraining set shape:", X_train_scaled.shape)
print("Test set shape:", X_test_scaled.shape)
print("\nClass distribution in training set:")
print(pd.Series(y_train).value_counts())
print("\nClass distribution in test set:")
print(pd.Series(y_test).value_counts())
```

```python
# Train logistic regression on imbalanced data
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train_scaled, y_train)

# Predictions
y_pred = lr.predict(X_test_scaled)
y_prob = lr.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

print("Baseline Model (Imbalanced Data) Performance:")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC: {auc:.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - Baseline Model')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Apply random oversampling
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(X_train_scaled, y_train)

print("Class distribution after oversampling:")
print(pd.Series(y_ros).value_counts())

# Train model
lr_ros = LogisticRegression(random_state=42, max_iter=1000)
lr_ros.fit(X_ros, y_ros)

# Predictions
y_pred_ros = lr_ros.predict(X_test_scaled)
y_prob_ros = lr_ros.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy_ros = accuracy_score(y_test, y_pred_ros)
f1_ros = f1_score(y_test, y_pred_ros)
auc_ros = roc_auc_score(y_test, y_prob_ros)

print("\nRandom Oversampling Model Performance:")
print(f"Accuracy: {accuracy_ros:.4f}")
print(f"F1 Score: {f1_ros:.4f}")
print(f"AUC: {auc_ros:.4f}")

# Confusion matrix
cm_ros = confusion_matrix(y_test, y_pred_ros)
plt.figure(figsize=(6,4))
sns.heatmap(cm_ros, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - Random Oversampling')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_ros))
```

```python
# Apply random undersampling
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X_train_scaled, y_train)

print("Class distribution after undersampling:")
print(pd.Series(y_rus).value_counts())

# Train model
lr_rus = LogisticRegression(random_state=42, max_iter=1000)
lr_rus.fit(X_rus, y_rus)

# Predictions
y_pred_rus = lr_rus.predict(X_test_scaled)
y_prob_rus = lr_rus.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy_rus = accuracy_score(y_test, y_pred_rus)
f1_rus = f1_score(y_test, y_pred_rus)
auc_rus = roc_auc_score(y_test, y_prob_rus)

print("\nRandom Undersampling Model Performance:")
print(f"Accuracy: {accuracy_rus:.4f}")
print(f"F1 Score: {f1_rus:.4f}")
print(f"AUC: {auc_rus:.4f}")

# Confusion matrix
cm_rus = confusion_matrix(y_test, y_pred_rus)
plt.figure(figsize=(6,4))
sns.heatmap(cm_rus, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - Random Undersampling')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rus))
```

```python
# Apply Tomek Links
tomek = TomekLinks()
X_tomek, y_tomek = tomek.fit_resample(X_train_scaled, y_train)

print("Class distribution after Tomek Links:")
print(pd.Series(y_tomek).value_counts())

# Train model
lr_tomek = LogisticRegression(random_state=42, max_iter=1000)
lr_tomek.fit(X_tomek, y_tomek)

# Predictions
y_pred_tomek = lr_tomek.predict(X_test_scaled)
y_prob_tomek = lr_tomek.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy_tomek = accuracy_score(y_test, y_pred_tomek)
f1_tomek = f1_score(y_test, y_pred_tomek)
auc_tomek = roc_auc_score(y_test, y_prob_tomek)

print("\nTomek Links Model Performance:")
print(f"Accuracy: {accuracy_tomek:.4f}")
print(f"F1 Score: {f1_tomek:.4f}")
print(f"AUC: {auc_tomek:.4f}")

# Confusion matrix
cm_tomek = confusion_matrix(y_test, y_pred_tomek)
plt.figure(figsize=(6,4))
sns.heatmap(cm_tomek, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - Tomek Links')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_tomek))
```

```python
# Apply SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train_scaled, y_train)

print("Class distribution after SMOTE:")
print(pd.Series(y_smote).value_counts())

# Train model
lr_smote = LogisticRegression(random_state=42, max_iter=1000)
lr_smote.fit(X_smote, y_smote)

# Predictions
y_pred_smote = lr_smote.predict(X_test_scaled)
y_prob_smote = lr_smote.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy_smote = accuracy_score(y_test, y_pred_smote)
f1_smote = f1_score(y_test, y_pred_smote)
auc_smote = roc_auc_score(y_test, y_prob_smote)

print("\nSMOTE Model Performance:")
print(f"Accuracy: {accuracy_smote:.4f}")
print(f"F1 Score: {f1_smote:.4f}")
print(f"AUC: {auc_smote:.4f}")

# Confusion matrix
cm_smote = confusion_matrix(y_test, y_pred_smote)
plt.figure(figsize=(6,4))
sns.heatmap(cm_smote, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - SMOTE')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_smote))
```

```python
# Compute class weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = {0: class_weights[0], 1: class_weights[1]}

# Train model with class weights
lr_weighted = LogisticRegression(random_state=42, max_iter=1000, class_weight=class_weights_dict)
lr_weighted.fit(X_train_scaled, y_train)

# Predictions
y_pred_weighted = lr_weighted.predict(X_test_scaled)
y_prob_weighted = lr_weighted.predict_proba(X_test_scaled)[:, 1]

# Metrics
accuracy_weighted = accuracy_score(y_test, y_pred_weighted)
f1_weighted = f1_score(y_test, y_pred_weighted)
auc_weighted = roc_auc_score(y_test, y_prob_weighted)

print("\nClass Weighting Model Performance:")
print(f"Accuracy: {accuracy_weighted:.4f}")
print(f"F1 Score: {f1_weighted:.4f}")
print(f"AUC: {auc_weighted:.4f}")

# Confusion matrix
cm_weighted = confusion_matrix(y_test, y_pred_weighted)
plt.figure(figsize=(6,4))
sns.heatmap(cm_weighted, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix - Class Weighting')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred_weighted))
```

```python
# Create comparison table
results = pd.DataFrame({
    'Technique': ['Imbalanced', 'Random Oversampling', 'Random Undersampling',
                 'Tomek Links', 'SMOTE', 'Class Weighting'],
    'Accuracy': [accuracy, accuracy_ros, accuracy_rus,
                 accuracy_tomek, accuracy_smote, accuracy_weighted],
    'F1 Score': [f1, f1_ros, f1_rus,
                 f1_tomek, f1_smote, f1_weighted],
    'AUC': [auc, auc_ros, auc_rus,
            auc_tomek, auc_smote, auc_weighted]
})

# Display results
print("Performance Comparison:")
display(results)

# Plot comparison
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.barplot(x='Technique', y='Accuracy', data=results)
plt.xticks(rotation=45)
plt.title('Accuracy Comparison')

plt.subplot(1, 3, 2)
sns.barplot(x='Technique', y='F1 Score', data=results)
plt.xticks(rotation=45)
plt.title('F1 Score Comparison')

plt.subplot(1, 3, 3)
sns.barplot(x='Technique', y='AUC', data=results)
plt.xticks(rotation=45)
plt.title('AUC Comparison')

plt.tight_layout()
plt.show()
```
